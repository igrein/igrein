```python
# Устанавливаем необходимые библиотеки
!pip install gdown

import gdown
import joblib
import numpy as np
import warnings
from sklearn.datasets import make_classification

# Загружаем модель с Google Диска
file_id_rf = '14-8_rLHC51qc96KUpQqg_AxplzyDPGs7'
url_rf = f'https://drive.google.com/uc?id={file_id_rf}'
gdown.download(url_rf, 'random_forest_model.joblib', quiet=False)

# Загружаем модель PCA с Google Диска
file_id_pca = '1Hpco6SMqjRfOqRIHiHmE6zXCk7_32zTw'
url_pca = f'https://drive.google.com/uc?id={file_id_pca}'
gdown.download(url_pca, 'pca_model.joblib', quiet=False)

# Загружаем модели с помощью joblib
rf_model = joblib.load('random_forest_model.joblib')
pca_model = joblib.load('pca_model.joblib')

print("Модели успешно загружены!")

```

```python
# Функция для получения ответа и рекомендаций
def get_prediction(user_input_data):
    # Преобразуем данные пользователя в формат, подходящий для модели
    input_data = np.array(user_input_data).reshape(1, -1)

    # Применяем PCA
    input_data_pca = pca_model.transform(input_data)

    # Получаем предсказание от модели
    prediction = rf_model.predict(input_data_pca)

    # Интерпретируем предсказание
    if prediction == 1:
        recommendation = "Молекула готова к биологическому разложению. Рекомендуется использовать в экологических проектах."
    else:
        recommendation = "Молекула не готова к биологическому разложению. Рекомендуется провести дополнительные исследования или изменить состав молекулы."

    # Возвращаем предсказание и рекомендацию
    return prediction, recommendation

# Пример ввода данных пользователя (молекулярные дескрипторы)
# Можно поменять на необходимые значения (всего 41 дескриптор)
user_input_data = [0.2, 0.5, 0.7, 0.9, 0.4, 0.6, 0.3, 0.1, 0.7, 0.8, 0.2, 0.6, 0.9, 0.4, 0.5, 0.7, 0.1, 0.8, 0.3, 0.9,
                   0.5, 0.4, 0.3, 0.6, 0.2, 0.4, 0.5, 0.7, 0.9, 0.1, 0.8, 0.6, 0.7, 0.2, 0.6, 0.7, 0.8, 0.4, 0.5, 0.6, 0.9]

#Получаем результат
prediction, recommendation = get_prediction(user_input_data)

# Выводим предсказание и рекомендации
print(f"Предсказание модели: {prediction}")
print(f"Рекомендация: {recommendation}")
```
